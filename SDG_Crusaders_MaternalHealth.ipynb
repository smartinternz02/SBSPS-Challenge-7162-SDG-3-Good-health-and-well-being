```python


x = ind_spec_copy1[" Women literacy(%)"]
y1 = ind_spec_copy1[" Mothers having at least 4 antenatal care visits (%)"]
m1, b1 = np.polyfit(x, y1, 1)

y2 = ind_spec_copy1["mothers or pregnant with age 15-19 at the time of the survey (%)"]
m2, b2 = np.polyfit(x, y2, 1)

y3 = ind_spec_copy1["Births in the 5 years preceding the survey that are third or higher order (%)"]
m3, b3 = np.polyfit(x, y3, 1)

y4 = ind_spec_copy1["pregnant and consumed iron folic acid for 180 days or more (%)"]
m4, b4 = np.polyfit(x, y4, 1)

fig, axs = plt.subplots(4, 1, sharex=True, figsize=(10,20))
axs[0].scatter(x, y1)
axs[0].set_xlabel("Women literacy(%)")
axs[0].set_ylabel("Mothers having at least 4 antenatal care visits(%)")
axs[0].plot(x, m1*x + b1)

axs[1].scatter(x, y2)
axs[1].set_xlabel("Women literacy(%)")
axs[1].set_ylabel("mothers or pregnant(age 15-19) at time of survey(%)")
axs[1].plot(x, m2*x + b2)

axs[2].scatter(x, y3)
axs[2].set_xlabel("Women literacy(%)")
axs[2].set_ylabel("Births 5 years preceding survey that are 3rd or higher order(%)")
axs[2].plot(x, m3*x + b3)

axs[3].scatter(x, y4)
axs[3].set_xlabel("Women literacy(%)")
axs[3].set_ylabel("pregnant and consumed iron folic acid for >=180 days(%)")
axs[3].plot(x, m4*x + b4)


#plt.tight_layout()
plt.show()
```

The above graphs tell us that as the literacy rate in women increases, the number of women who are pregnant or mothers at an age of 15-19 years, and women who have 3 or more children, decreases. This implies that increased literacy in women can possibly lead to improved family planning and reduce pregnancies at a young age for women.
The above graphs also show that as the literacy rate in women increases, there is also an increase in women who opt to go for atleast 4 antenatal care visits and those that have consumed iron folic acid for more than 180 days during pregnancy. 
This implies that an increased literacy rate can also possibly increase awareness in women and lead more women to opt for necessary maternal healthcare.

```python
x = ind_spec_copy1["Births attended by skilled health personnel(%)"]
y1 = ind_spec_copy1["Institutional births in public facility (%)"]
m1, b1 = np.polyfit(x, y1, 1)
plt.scatter(x, y1)
#plt.set_xlabel("Women literacy(%)")
#plt.set_ylabel("Mothers having at least 4 antenatal care visits(%)")
plt.title("Births attended by skilled health personnel(%) Vs Institutional births in public facility (%)")
plt.plot(x, m1*x + b1)
plt.show()
```

```python
x = ind_spec_copy1["Registered pregnancies that received (MCP) card (%)"]
y1 = ind_spec_copy1["Institutional births in public facility (%)"]
m1, b1 = np.polyfit(x, y1, 1)
plt.scatter(x, y1)
#plt.set_xlabel("Women literacy(%)")
#plt.set_ylabel("Mothers having at least 4 antenatal care visits(%)")
plt.title("Institutional births in public facility (%) Vs Registered pregnancies that received (MCP) card")
plt.plot(x, m1*x + b1)
plt.show()
```

The two plots that compare births in public healthcare facilitites with registered pregnancies that received Mother and Child Protection cards, and with the number of births attended by skilled health personnel shows that in districts where more skilled health personnel have attended births, the number of births taking place in public healthcare facilities is slightly higher.
The above graph also shows the possibility that districts where more pregnant women received mother and Child Protection cards had more births took place in public healthcare facilities.
What we can infer from this is that when adequate facilities and resources are provided by the government, the likeliness of mothers preferring public healthcare increases. 

```python
x = ind_spec_copy1["Population living with improved sanitation facility (%)"]
y1 = ind_spec_copy1["Women(15-24)using hygienic methods for mensuration(%)"]
m1, b1 = np.polyfit(x, y1, 1)
plt.scatter(x, y1)
#plt.set_xlabel("Women literacy(%)")
#plt.set_ylabel("Mothers having at least 4 antenatal care visits(%)")
plt.title("Population living with improved sanitation facility Vs Women(15-24)using hygienic methods for menstruation")
plt.plot(x, m1*x + b1)
plt.show()
```
